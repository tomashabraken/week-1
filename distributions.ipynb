```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('class_data.csv')
df.head()
```

```python
df_insta = df[df['instagram_followers']<500]
```

```python
sns.boxplot(df_insta['instagram_followers'])
plt.title('Instagam followers')
plt.xlabel('Number of followers')
plt.show()
```

```python
df['linkedin_con'].name = 'linkedin connecties' #Labeling the variable for plotting
linkedin = df['linkedin_con'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(linkedin, kde=False) 
plt.title('Number of connections')
plt.show()
```

```python

mean_con = df['linkedin_con'].mean()
median_con = df['linkedin_con'].median()

print(f'The median linkedin connecties is: {median_con} connecties')
print(f'The mean linkedin connecties is: {mean_con} connecties')

#note: need to copy code
sns.distplot(linkedin, kde=False) #I need to drop the NA's, otherwise Seaborn won't plot
plt.title('linkdin connections')
plt.axvline(median_con, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_con, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
df_apps = df[df['apps_phone'] <2000] #Subsetting the dataframe to include only paid apps
sns.violinplot(df_apps['apps_phone'])
sns.swarmplot(df_apps['apps_phone'], color='yellow') #default color doesn't stand out
plt.title('Number of apps')
plt.ylabel('Frequency')
plt.xlabel('Apps')
plt.show()
```
