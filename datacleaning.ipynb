```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("steps_data.csv", sep=';')
df1.head()
```

```python
df2 = pd.read_csv("survey_data.csv")
df2.head()
```

```python
df_merge = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df_merge.head()
```

```python
df_merge['waist'].value_counts()
```

```python
# Python would normally generate. 
df_merge['waist'] = df_merge['waist'].apply(pd.to_numeric, errors='coerce')
#To remove the values above 5 we need to create a function that we then apply to the entire column
def above_110(x):
    if(x > 110.0): 
        return float('NaN')
    else: 
        return x
df_merge['waist'] = df_merge['waist'].apply(above_110)
```

```python
def under_65(x):
    if(x < 65.0): 
        return float('NaN')
    else: 
        return x
df_merge['waist'] = df_merge['waist'].apply(under_65)
```
